# Sampling a signal

```python
sound.normalize()
```

```python
import numpy as np
def sample(wave, factor):
   ys = np.zeros(len(wave))
   ys[::factor] = wave.ys[::factor]
   return thinkdsp.Wave(ys, framerate=wave.framerate)
```

```python
# Sampling
def make_impulses(wave, factor):
  ys = np.zeros(len(wave))
  ys[::factor] = 1
  ts = np.arange(len(wave)) / wave.framerate
  return thinkdsp.Wave(ys, ts, wave.framerate)
```

```python
sampled_sound_wave=make_impulses(sound, 2000)*sound
sampled_sound_wave.plot()
```

```python
sampled_sound_wave.write('sampledS1.wav')
sampled_sound_wave.make_audio()
```
